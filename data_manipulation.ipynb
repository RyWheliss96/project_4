```python
import numpy as np
import pandas as pd
import warnings
from pathlib import Path

warnings.filterwarnings('ignore')
```

```python
file_path = Path("Data/spreadspoke_scores.csv")
df_big_data = pd.read_csv(file_path)

df_big_data.head()
```

```python
def new_col(row):
    if row['score_home'] == row['score_away']:
        val = 2
    elif row['score_home'] > row['score_away']:
        val = 1
    else:
        val = 0
    return val
```

```python
df_big_data['winner(0_away,1_home,2_tie)'] = df_big_data.apply(new_col, axis=1)

df_big_data.head()
```

```python
file_path2 = Path("Data/nfl_teams.csv")
df_teams = pd.read_csv(file_path2)
df_teams.head()
```

```python
file_path3 = Path("Data/stadiums.csv")
df_stadiums = pd.read_csv(file_path3, encoding = 'unicode_escape')
df_stadiums.head()
```

```python
bigoledata = df_big_data.merge(df_teams, left_on = 'team_home', right_on = 'team_name', how = 'left', suffixes = (None, '_home'))
bigoledata.head()
```

```python
humungodata = bigoledata.merge(df_teams, left_on = 'team_away', right_on = 'team_name', how = 'left', suffixes = ('_home', '_away'))
humungodata.head()
```

```python
gigadata = humungodata.merge(df_stadiums, left_on = 'stadium', right_on = 'stadium_name', how = 'left')
gigadata.head()
```

```python
gigadata.columns
```

```python
gigadata = gigadata.drop(columns =['team_home', 'score_home', 'score_away', 'team_away','stadium'] )
gigadata.columns
```

```python
gigadata['schedule_date'] = gigadata['schedule_date'].str.replace(r'\D', '', regex=True).astype(int)
gigadata.head()
```

```python
for index,row in gigadata.iterrows():
    if row['schedule_playoff']==True:
        gigadata['schedule_week'][index] = 20
        gigadata['schedule_playoff'][index]=1
    else:
        gigadata['schedule_playoff'][index]=0

gigadata.head()
```

```python
print(df_teams[df_teams['team_id']=='LAR'].index.values)
```

```python
for index,row in gigadata.iterrows():
    if pd.isnull(row['team_favorite_id']) or row['team_favorite_id']=='PICK'or row['team_favorite_id']=='LV':
        gigadata['team_favorite_id'][index] = -1
    elif row['team_favorite_id'] in ['ARI','IND','NE','TEN','LVR','LAC','LAR','WAS']:
        if row['team_id_home']==row['team_favorite_id']:
            gigadata['team_favorite_id'][index]= df_teams[df_teams['team_name']==row['team_name_home']].index.values
        else:
            gigadata['team_favorite_id'][index]= df_teams[df_teams['team_name']==row['team_name_away']].index.values
    else:
        gigadata['team_favorite_id'][index]= df_teams[df_teams['team_id']==row['team_favorite_id']].index.values

gigadata.head()
```

```python
for index,row in gigadata.iterrows():
    if pd.isnull(row['spread_favorite']):
        gigadata['spread_favorite'][index] = -1
```

```python
print(gigadata['over_under_line'][2728])
```

```python
for index,row in gigadata.iterrows():
    if pd.isnull(row['over_under_line']) or row['over_under_line']==' ':
        gigadata['over_under_line'][index] = -1
```

```python
for index,row in gigadata.iterrows():
    if row['stadium_neutral'] == True:
        gigadata['stadium_neutral'][index] = 1
    else:
        gigadata['stadium_neutral'][index] = 0
```

```python
weather_options = ['fog','indoor','rain','rain | fog', 'retractable (open roof)','snow','snow | fog','snow | Freezing rain']

for index,row in gigadata.iterrows():
    if row['weather_detail'] in weather_options:
        gigadata['weather_detail'][index]= weather_options.index(row['weather_detail'])
    else:
        gigadata['weather_detail'][index]= -1
gigadata.head()
```

```python
gigadata = gigadata.drop(columns =['team_name_short_home', 'team_id_home','team_id_pfr_home', 'team_name_short_away', 'team_id_away','team_id_pfr_away', 'stadium_location','stadium_address'])
```

```python
for index,row in gigadata.iterrows():
    gigadata['team_name_home'][index]= df_teams[df_teams['team_name']==row['team_name_home']].index.values
    gigadata['team_name_away'][index]= df_teams[df_teams['team_name']==row['team_name_away']].index.values
```

```python
for index,row in gigadata.iterrows():
    if row['team_conference_home']=='AFC':
        gigadata['team_conference_home'][index]=0
    else:
        gigadata['team_conference_home'][index]=1
```

```python
for index,row in gigadata.iterrows():
    if row['team_conference_pre2002_home']=='AFC':
        gigadata['team_conference_pre2002_home'][index]=0
    else:
        gigadata['team_conference_pre2002_home'][index]=1
```

```python
for index,row in gigadata.iterrows():
    if row['team_conference_away']=='AFC':
        gigadata['team_conference_away'][index]=0
    else:
        gigadata['team_conference_away'][index]=1
```

```python
for index,row in gigadata.iterrows():
    if row['team_conference_pre2002_away']=='AFC':
        gigadata['team_conference_pre2002_away'][index]=0
    else:
        gigadata['team_conference_pre2002_away'][index]=1
```

```python
conferences = ['AFC North', 'AFC South', 'AFC East', 'AFC West','AFC Central','NFC North', 'NFC South','NFC East','NFC West','NFC Central']
for index,row in gigadata.iterrows():
    if row['team_division_home'] in conferences:
        gigadata['team_division_home'][index]= conferences.index(row['team_division_home'])
    else:
        gigadata['team_division_home'][index]=-1
for index,row in gigadata.iterrows():
    if row['team_division_pre2002_home'] in conferences:
        gigadata['team_division_pre2002_home'][index]= conferences.index(row['team_division_pre2002_home'])
    else:
        gigadata['team_division_pre2002_home'][index]=-1
```

```python
for index,row in gigadata.iterrows():
    if row['team_division_away'] in conferences:
        gigadata['team_division_away'][index]= conferences.index(row['team_division_away'])
    else:
        gigadata['team_division_away'][index]=-1
for index,row in gigadata.iterrows():
    if row['team_division_pre2002_away'] in conferences:
        gigadata['team_division_pre2002_away'][index]= conferences.index(row['team_division_pre2002_away'])
    else:
        gigadata['team_division_pre2002_away'][index]=-1
```

```python
for index,row in gigadata.iterrows():
    gigadata['stadium_name'][index]= df_stadiums[df_stadiums['stadium_name']==row['stadium_name']].index.values
```

```python
type_stadium = ['indoor','outdoor','retractable']

for index,row in gigadata.iterrows():
    if row['stadium_type']in type_stadium:
        gigadata['stadium_type'][index]= type_stadium.index(row['stadium_type'])
    else:
        gigadata['stadium_type'][index]=-1
```

```python
weather_type = ['warm','cold','moderate','indoor']

for index,row in gigadata.iterrows():
    if row['stadium_weather_type']in type_stadium:
        gigadata['stadium_weather_type'][index]= weather_type.index(row['stadium_weather_type'])
    else:
        gigadata['stadium_weather_type'][index]=-1
```

```python
surface = ['Grass','FieldTurf','Grass, Turf (1969-1970)','Grass, Turf (1970-1971)', 'Grass, Turf (1971-1974)', 'Hellas Matrix Turf']

for index,row in gigadata.iterrows():
    if row['stadium_surface']in type_stadium:
        gigadata['stadium_surface'][index]= surface.index(row['stadium_surface'])
    else:
        gigadata['stadium_surface'][index]=-1

gigadata.head()
```

```python
gigadata.to_csv("Data/final_data.csv")
```
